**Breadth First Search:**
explores the tree level by level, making sure each node is traversed within that level,
search continues until goal node found or tree finished.
**Tree Structure:** The tree is represented as a dictionary, where each key is a parent
node, and the value is a list of its children.

BFS(graph,start,goal):
The bfs function performs a breadth-first search on the given tree (graph) starting from
the start node and searches for the goal node.

1) make an empty visited list to track visited nodes
2) queue(fifo) to manage the nodes to be explored
3) starting node added to both queue and visited
4) continue the search until all nodes in queue are there
    . dequeue the first node in the queue 
    . if node=goal print node
    . enqueue all unvisited neighbours of curr node
    

```python
graph = {
    "A": ["B","C"],
    "B": ["D","E"],
    "C": ["F","G"],
    "D": ["H"],
    "E": [],
    "F": ["I"],
    "G": [],
    "H": [],
    "I": []
}

def BFS(graph,start,goal):
    visited = []
    queue = [[start]]
    
    while queue:
        path = queue.pop(0)
        node = path[-1] 
        print(node)

        if node == goal:
            print("goal found!","->".join(path))
            return path
            break
        if node not in visited:
            visited.append(node)

        for neighbour in graph[node]:
            new_path = path + [neighbour]
            queue.append(new_path)

start = "A"
goal = "I"
BFS(graph,start,goal)




```

MAZE REPRESENTATION:
The given Python code represents a maze as a graph and utilizes Breadth-First Search
(BFS) to find a path from the start node (0,0) to the goal node (2,2). The maze is a 2D
grid where 1 represents an open path and 0 represents a blocked path. 
The create_graph() function constructs a graph using an adjacency list, where each open cell
is treated as a node, and edges connect nodes that have valid right or down movements.
Once the graph is built, the bfs() function performs BFS traversal by exploring nodes
layer by layer, ensuring the shortest path is found. A queue is used to process nodes in
FIFO order, marking visited nodes to avoid repetition. The traversal starts from (0,0),visiting all reachable neighbors until the goal (2,2) is found, printing the traversal path
along the way.

```python
maze= [
    [1,1,0],
    [0,1,0],
    [0,1,1]
]

directions = [(0,1),(1,0)] #row,col

#comvertn maze to graph(adjacency lists):
def create_graph(maze):
    graph = {}
    rows = len(maze) # 3 rows
    col = len(maze[0]) # 3 cols

    for i in range(rows):
        for j in range(col):
            if maze[i][j] == 1: #if an open path
                neighbours = []
                for dx,dy in directions:
                    nx,ny = i + dx, j+dy
                    if 0<=nx<rows and 0<=ny<col and maze[nx][ny] == 1:
                        neighbours.append((nx,ny))
                graph[(i,j)] = neighbours
    return graph

def bfs(graph,start,goal):
    visited = []
    queue = [[start]]

    while queue:
        path = queue.pop(0)
        node = path[-1]

        if node == goal:
             return f"Goal found: {' -> '.join(map(str, path))}"
        
        if node not in visited:
            visited.append(node)

        for neighbour in graph.get(node,[]):
            if neighbour not in visited:
                new_path = path + [neighbour]
                queue.append(new_path)
        
    return "not found"

graph = create_graph(maze)
start = (0,0)
goal = (2,2)
print(bfs(graph,start,goal))


```

DEPTH FIRST SEARCH:
to traverse a tree like structure
starts at a given node and explores as deep as possible through that node, search continues until goal found or all nodes have been traversed.

DFS(graph,start,goal):
1) visited list to keep track of all visited nodes
2) stack(LIFO) store nodes that have been explored, DFS is LIFO explored nodes are depth-first order
3) Push start node in stack to keep track of visited:
    . enter a loop where it pops nodes fromnstack and      explores them
        . if popped node = goal, search stops
        . if popped node has unvisited neighbours, they are pushed to stack for further exploration
        . if popped has no neighbours, then we bcktrack by popping next node from stack

```python
def dfs(graph,start,goal):
    visited = []
    stack = []

    visited.append(start)
    stack.append(start)

    while stack:
        node = stack.pop() #LIFO pop from top(last visited node) in this case the first node
        print(node,end=" ")

        if node == goal:
            print("found")
            break

        for neighbour in reversed(graph.get(node,[])):
            if neighbour not in visited:
                visited.append(neighbour)
                stack.append(neighbour)

graph = {
    "A": ["B","C"],
    "B": ["D","E"],
    "C": ["F","G"],
    "D": ["H"],
    "E": [],
    "F": ["I"],
    "G": [],
    "H": [],
    "I": []
}
start = "A"
goal = "I"
dfs(graph,start,goal)
```

A* search:
1) create an open list(priority queue) and closed list (visited nodes)
2) add start node to open list
3) **LOOP UNTIL GOAL FOUND** 
    . pick node with the lowest f(n) value from open list
    . move it close list
    . if goal , stop
    . else, generate neghbours
        . **NEIGHBOURS::**
            calculate g(n),h(n) and f(n)
                if neighbour not in open/closed or has lower cost, update it.            

```python
from queue import PriorityQueue
graph = {
    'A': {'B': 2, 'C': 1},
    'B': {'D': 4, 'E': 3},
    'C': {'F': 1, 'G': 5},
    'D': {'H': 2},
    'E': {},
    'F': {'I': 6},
    'G': {},
    'H': {},
    'I': {} 
}

heuristic = {
    "A": 7 , "B":6, "C":5, "D":4, "E":7, "F":3, "G":6, 
    "H":2, "I":0
}

def a_star(graph,start,goal):
    frontier = PriorityQueue()
    frontier.put((0+heuristic[start],start)) #f(n),node

    visited = set()
    g_costs = {start:0}
    came_from = {start:None}

    while not frontier.empty():
        _,curr_node = frontier.get()

        if curr_node in visited:
            continue

        #print(curr_node, end=" ")
        visited.add(curr_node)

        if curr_node == goal:
            path = []
            while curr_node:
                path.append(curr_node)
                curr_node = came_from[curr_node]
            path.reverse()
            print("found:","->".join(path))
            return path
        
        for neighbour, cost in graph.get(curr_node,{}).items():
            new_g_cost = g_costs[curr_node]+cost
            f_cost = new_g_cost + heuristic[neighbour]

            if neighbour not in g_costs or new_g_cost < g_costs[neighbour]:
                g_costs[neighbour] = new_g_cost
                came_from[neighbour] = curr_node
                frontier.put((f_cost,neighbour))
    print("not found")
    return None

print("A*: ")
a_star(graph,"A","I")
```
