Practice Task # 1 (Model-Based Agent)
Scenario:
A smart greenhouse monitors and regulates plant growth by controlling humidity and temperature. The Greenhouse Agent is responsible for detecting environmental conditions and adjusting settings to ensure optimal plant growth.
Task:
Initial Check:
The environment is initialized with random temperature and humidity levels.
Display the initial state of the greenhouse.
Monitoring & Adjustment:
The agent scans the environment.
If the temperature is too high, the agent turns on a cooling system.
If the humidity is too low, the agent activates the misting system.
If conditions are optimal, no action is taken.
Final System Check:
Display the updated greenhouse conditions after adjustments.


```python
import random
class Environment:
    def __init__(self):
        self.temp = random.choice(["High","Low"])
        self.humidity = random.choice(["High","Low"])
    
    def initial(self):
        return {"temp": self.temp , "humidity": self.humidity}
    
    def updated_state(self,new_temp=None , new_humidity=None):
        if new_temp:
            self.temp=new_temp
        if new_humidity:
            self.humidity=new_humidity
    
class GHA:
    def __init(self):
        self.model = {}
    
    def act(self,state,environment):
        self.model = state
        actions = []

        if state["temp"] == "High":
            actions.append("Cooliing system turned on")
            environment.updated_state(new_temp="OK")
        if state["humidity"] == "Low":
            actions.append("Misting turned on")
            environment.updated_state(new_humidity="OK")
        if not actions:
            print("no action taken")

        return actions

def run_agent(environment,gha):
    ini = environment.initial()
    print(f"initial: {ini}")

    action = gha.act(ini,environment)

    final = environment.initial()
    print(f"actions: {action} \n Update: {final}")

environment=Environment()
gha = GHA()

run_agent(environment,gha)
```

Practice Task # 2 (Utility-Based Agent)
Scenario:
A self-driving taxi operates in a city and must decide the best route for passengers. The Taxi Agent must select the most efficient path based on traffic conditions, distance, and fuel consumption.
Task:
Initial Setup:
The city is represented as a grid with multiple routes.
Each route has attributes like traffic congestion level, distance, and fuel cost.
Display available routes.
Route Selection:
The agent evaluates each route using a utility function that balances shortest time, lowest fuel cost, and minimum congestion.
The agent selects the best route.
Final Decision:
Display the chosen route and its expected benefits.

```python
class Environment:
    def __init__(self):
        self.routes = {
            "North": ["high",10,4], #traffic,dist,fuel
            "South": ["no", 7,6],
            "East": ["no",6,3],
            "West": ["high", 10 , 10]
        }
    def get_percept(self):
        return self.routes

class Utility:
    def __init__(self):
        pass
    
    def act(self,percept,environment):
        optimal_path = None
        optimal_details = None
        min_dist = float("inf")
        min_fuel = float("inf")

        for route,details in percept.items():
            traffic,dist,fuel = details

            if traffic == "no" and (dist<min_dist or (dist==min_dist and fuel<min_fuel)):
                optimal_path = route
                min_dist=dist
                min_fuel=fuel
                optimal_details = details

        return optimal_path,optimal_details
    
def run_agent(environment,uti):
    
    percept = environment.get_percept()
    print(f"Initial: {percept}")

    opt = uti.act(percept,environment)
    print(f"Optimal: {opt}")

environment = Environment()
uti = Utility()

run_agent(environment,uti)
```

Practice Task # 3 (Goal-Based Agent)
Scenario:
A warehouse robot is responsible for picking and delivering packages. The Robot Agent must navigate between storage shelves and delivery points while optimizing efficiency.
Task:
Environment Setup:
The warehouse is represented as a grid.
Packages are placed at random locations.
Delivery points are predefined.
Navigation & Delivery:
The agent scans for the nearest package.
It moves to the packageâ€™s location, picks it up, and delivers it to the correct spot.
If there are multiple packages, the agent follows an optimized delivery path.
Final Check:
Display the final warehouse state with all packages delivered.

```python
import random

class Warehouse:
    def __init__(self, grid_size=5, num_packages=3):
        self.grid_size = grid_size  # Warehouse size (grid_size x grid_size)
        self.robot_position = (0, 0)  # Robot starts at (0,0)
        
        # Generate random package locations
        self.packages = {}
        while len(self.packages) < num_packages:
            location = (random.randint(0, grid_size-1), random.randint(0, grid_size-1))
            if location != self.robot_position:  # Ensure package is not at the start
                self.packages[location] = f"Package-{len(self.packages) + 1}"
        
        # Set predefined delivery points (e.g., bottom-right corner)
        self.delivery_point = (grid_size-1, grid_size-1)

    def get_state(self):
        """Returns the current warehouse state (robot position, packages, and delivery point)."""
        return {
            "robot_position": self.robot_position,
            "packages": self.packages,
            "delivery_point": self.delivery_point
        }

    def move_robot(self, new_position):
        """Moves the robot to a new position."""
        self.robot_position = new_position

    def pick_package(self, position):
        """Removes a package from the grid when picked up."""
        if position in self.packages:
            return self.packages.pop(position)
        return None

class GoalBasedRobot:
    def __init__(self):
        self.goal = "Deliver all packages"

    def find_nearest_package(self, robot_position, packages):
        """Finds the nearest package to the robot based on Manhattan distance."""
        min_distance = float("inf")
        nearest_package = None
        for position in packages.keys():
            distance = abs(robot_position[0] - position[0]) + abs(robot_position[1] - position[1])  # Manhattan distance
            if distance < min_distance:
                min_distance = distance
                nearest_package = position
        return nearest_package

    def act(self, warehouse):
        """Executes the robot's task of picking up and delivering packages."""
        state = warehouse.get_state()
        robot_position = state["robot_position"]
        packages = state["packages"]
        delivery_point = state["delivery_point"]

        actions = []

        while packages:
            nearest_package = self.find_nearest_package(robot_position, packages)
            if not nearest_package:
                break  # No more packages to pick up

            # Move to the package location
            actions.append(f"Moving to package at {nearest_package}")
            warehouse.move_robot(nearest_package)

            # Pick up the package
            package = warehouse.pick_package(nearest_package)
            actions.append(f"Picked up {package}")

            # Move to delivery point
            actions.append(f"Delivering {package} to {delivery_point}")
            warehouse.move_robot(delivery_point)
            actions.append(f"Delivered {package}")

            # Update robot position
            robot_position = delivery_point
        
        return actions

def run_agent():
    warehouse = Warehouse(grid_size=5, num_packages=3)
    robot = GoalBasedRobot()

    # Initial state
    print("\nInitial Warehouse State:")
    print(f"Robot Position: {warehouse.robot_position}")
    print(f"Packages: {warehouse.packages}")
    print(f"Delivery Point: {warehouse.delivery_point}")

    # Run the agent
    actions = robot.act(warehouse)

    # Print actions taken
    print("\nActions Taken:")
    for action in actions:
        print(action)

    # Final state
    print("\nFinal Warehouse State:")
    print(f"Robot Position: {warehouse.robot_position}")
    print(f"Packages Remaining: {warehouse.packages}")

# Run the warehouse simulation
run_agent()
```
