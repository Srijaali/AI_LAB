```python
# csp_ga_examples.py
# Examples of Constraint Satisfaction Problems (CSP) using OR-Tools and a simple Genetic Algorithm (GA)

# ----------- CSP Examples -----------
from ortools.sat.python import cp_model

def csp_examples():
    print("CSP Examples:")
    model = cp_model.CpModel()
    num_vals = 3

    # Declare variables x, y, z with domains {0,1,2}
    x = model.new_int_var(0, num_vals - 1, "x")
    y = model.new_int_var(0, num_vals - 1, "y")
    z = model.new_int_var(0, num_vals * num_vals, "z")  # Adjusted to accommodate product

    # Unary constraint: x != 2
    model.add(x != 2)

    # Binary constraint: x != y
    model.add(x != y)

    # Global constraint: all variables must take different values
    model.add_all_different([x, y])  # z is removed since it might be product of x and y

    # Linear constraint: x + y <= 3
    model.add(x + y <= 3)

    # Non-linear constraint approximation: x * y == z using AddMultiplicationEquality
    model.add_multiplication_equality(z, [x, y])

    # Solve the model
    solver = cp_model.CpSolver()
    status = solver.solve(model)

    if status in (cp_model.OPTIMAL, cp_model.FEASIBLE):
        print(f"Solution found: x={solver.value(x)}, y={solver.value(y)}, z={solver.value(z)}")
    else:
        print("No solution found.")


# ----------- Genetic Algorithm Example -----------
import random

def genetic_algorithm():
    print("\nGenetic Algorithm Example:")
    # GA parameters
    pop_size = 10
    chromosome_length = 5  # 5-bit binary string
    generations = 20
    crossover_rate = 0.7
    mutation_rate = 0.1

    # Fitness function: decode binary to integer and square it
    def decode(chrom):
        return int("".join(str(bit) for bit in chrom), 2)
    def fitness(chrom):
        x_val = decode(chrom)
        return x_val * x_val

    # Initialize random population
    population = [[random.randint(0, 1) for _ in range(chromosome_length)] for _ in range(pop_size)]

    for gen in range(generations):
        # Evaluate fitness for each individual
        fits = [fitness(ind) for ind in population]
        print(f"Generation {gen}: Best fitness = {max(fits)}")

        # Selection (tournament)
        new_pop = []
        while len(new_pop) < pop_size:
            # Select two parents via tournament
            a, b = random.sample(population, 2)
            parent1 = a if fitness(a) > fitness(b) else b
            a, b = random.sample(population, 2)
            parent2 = a if fitness(a) > fitness(b) else b

            # Crossover
            if random.random() < crossover_rate:
                point = random.randint(1, chromosome_length - 1)
                child1 = parent1[:point] + parent2[point:]
                child2 = parent2[:point] + parent1[point:]
            else:
                child1, child2 = parent1[:], parent2[:]

            # Mutation
            def mutate(chrom):
                return [bit ^ 1 if random.random() < mutation_rate else bit for bit in chrom]

            new_pop.extend([mutate(child1), mutate(child2)])

        population = new_pop[:pop_size]

    # Report best solution
    best = max(population, key=fitness)
    print(f"Best solution: value={decode(best)}, fitness={fitness(best)}")


if __name__ == "__main__":
    csp_examples()
    genetic_algorithm()
```

```python
# Task 1:
class Product:
    def __init__(self, name, frequency, volume):
        self.name = name
        self.frequency = frequency
        self.volume = volume
        self.efficiency = frequency / volume

class Slot:
    def __init__(self, id, distance):
        self.id = id
        self.distance = distance
        self.capacity = 1  # Assuming each slot holds one product
        self.assigned_product = None

products = [
    Product("Product 1", frequency=15, volume=2),
    Product("Product 2", frequency=8, volume=1),
    Product("Product 3", frequency=20, volume=3)
]

slots = [
    Slot("Slot 1", distance=1),
    Slot("Slot 2", distance=2),
    Slot("Slot 3", distance=3)
]

products.sort(key=lambda p: p.efficiency, reverse=True)

slots.sort(key=lambda s: s.distance)

assignments = {}
used_slots = set()

for product in products:
    for slot in slots:
        if slot.id not in used_slots:
            assignments[product.name] = slot.id
            used_slots.add(slot.id)
            break

print("Product Assignments (Closer slot = less walking distance):\n")
for product in products:
    assigned_slot = assignments.get(product.name, "Not Assigned")
    print(f"{product.name} → {assigned_slot}")
```

```python
#gpt
from ortools.sat.python import cp_model

# Product data: (Name, Frequency, Volume)
products = [("Product 1", 15, 2),
            ("Product 2", 8, 1),
            ("Product 3", 20, 3)]

# Slot data: (Slot ID, Distance from dispatch)
slots = [("Slot 1", 1),
         ("Slot 2", 2),
         ("Slot 3", 3)]

num_products = len(products)
num_slots = len(slots)

model = cp_model.CpModel()

# Create variables: assign[i] = slot index assigned to product i
assign = [model.new_int_var(0, num_slots - 1, f'assign_p{i}') for i in range(num_products)]

# Ensure each slot is used by only one product
model.add_all_different(assign)

# Map distance from slot index to a constant list
distances = [dist for (_, dist) in slots]
frequencies = [freq for (_, freq, _) in products]

# Compute total weighted walking distance: sum(freq[i] * dist[assign[i]])
total_distance_terms = []
for i in range(num_products):
    # Element constraint: distance[i] = distances[assign[i]]
    dist_var = model.new_int_var(min(distances), max(distances), f'distance_p{i}')
    model.add_element(assign[i], distances, dist_var)
    total_distance_terms.append(dist_var * frequencies[i])

model.Minimize(sum(total_distance_terms))

# Solve
solver = cp_model.CpSolver()
status = solver.solve(model)

# Output result
if status == cp_model.OPTIMAL or status == cp_model.FEASIBLE:
    print("\nOptimal Assignment (Minimizing Total Walking Distance):\n")
    for i in range(num_products):
        prod_name = products[i][0]
        assigned_slot_idx = solver.value(assign[i])
        slot_name = slots[assigned_slot_idx][0]
        print(f"{prod_name} → {slot_name}")
    print(f"\nTotal Weighted Walking Distance: {solver.objective_value}")
else:
    print("No solution found.")
```
